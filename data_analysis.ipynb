```python
import numpy as np
import matplotlib.pyplot as plt
import csv
  
entrada = open('data.csv')
tabla = []
for fila in csv.reader(entrada): 
    tabla.append(fila)  
entrada.close()

t = np.linspace(0, len(tabla),len(tabla)-1)
aax = []
aay = []
aaz = []
yaw = []
pitch = []
roll = []
temperature = []
pressure = []
altitude = []

for fila in range(1, len(tabla)):
    aax.append(float(tabla[fila][0]))  
    aay.append(float(tabla[fila][1]))
    aaz.append(float(tabla[fila][2]))
    yaw.append(float(tabla[fila][3]))  
    pitch.append(float(tabla[fila][4]))
    roll.append(float(tabla[fila][5]))
    temperature.append(float(tabla[fila][6]))  
    pressure.append(float(tabla[fila][7]))
    altitude.append(float(tabla[fila][8]))
```

```python
fig,ax = plt.subplots(5,1,figsize=(24,48))

ax1 = ax[0] # accel axis
cmap = plt.cm.Set1
ax1.plot(t,aax,label='x-Acceleration',color=cmap(0),linewidth=4)
ax1.plot(t,aay,label='y-Acceleration',color=cmap(1),linewidth=4)
ax1.plot(t,aaz,label='z-Acceleration',color=cmap(2),linewidth=4)
ax1.legend(fontsize=16)
ax1.grid()

ax2 = ax[1]
ax2.plot(t,yaw,label='yaw',color=cmap(0),linewidth=4)
ax2.plot(t,pitch,label='pitch',color=cmap(1),linewidth=4)
ax2.plot(t,roll,label='roll',color=cmap(2),linewidth=4)
ax2.legend(fontsize=16)
ax2.grid()

ax3 = ax[2]
ax3.plot(t,pressure,label='pressure',color=cmap(1),linewidth=4)
ax3.legend(fontsize=16)
ax3.grid()

ax4 = ax[3]
ax4.plot(t,temperature,label='temperature',color=cmap(0),linewidth=4)
ax4.legend(fontsize=16)
ax4.grid()

ax5 = ax[4]
ax5.plot(t,altitude,label='altitude',color=cmap(2),linewidth=4)
ax5.legend(fontsize=16)
ax5.grid()





```
